```python
import matplotlib.pyplot as plt
import numpy as np
import gdal
```

```python
f = gdal.Open('/Users/yuanhe/Documents/icloud_share/13_machine_deep_learning/02_SDD结果汇总20210601/代码/Code_summary(代码总结)/data/Lake_Chao_DGRN.tif')
in_band = f.GetRasterBand(1)
xsize = in_band.XSize
ysize = in_band.YSize
in_data = in_band.ReadAsArray()
mid = np.squeeze(in_data.reshape(-1,1))
lake_chao_sdd = []
for i in range(len(mid)):
    if mid[i] >= -10: 
        lake_chao_sdd.append(mid[i])
    else:
        pass
del i
lake_chao_sdd = np.array(lake_chao_sdd)
for x in range(lake_chao_sdd.shape[0]):
    if lake_chao_sdd[x,] < 0:
        lake_chao_sdd[x,] = np.nan

del mid,in_data,f,x
```
